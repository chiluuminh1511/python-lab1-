```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 16 20:00:29 2020

@author: chi
"""

#load libraries
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import eli5
from eli5.sklearn import PermutationImportance
#Import pandas,numpy for Dataset Manupilation and matplotlib and seaborn for Visualization and graphviz for tree visualization
import matplotlib.pyplot as plt 
import numpy as np
import graphviz

col_names='Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'

           #Load dataset
path='/home/chi/Desktop/anaconda-navigator/Lab 1/diabetes.csv'
df=pd.read_csv(path,header=0,names=col_names)
df.head()
```

```python
df.info()
feature_cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
features=feature_cols

X=df[feature_cols]#Features
y=df.Outcome#Target variable
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
clf=DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("Accuracy on Test Set:",metrics.accuracy_score(y_test, y_pred))
```

```python
perm=PermutationImportance(clf,random_state=1).fit(X_test,y_test)
eli5.show_weights(perm,feature_names=X_test.columns.tolist())

```
